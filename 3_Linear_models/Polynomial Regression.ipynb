```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression, Ridge, SGDRegressor, Lasso, ElasticNet

sns.set_style('darkgrid')
```

```python
m = 100
X = 6 * np.random.rand(m , 1) -3
y = 0.5 * X**2 + X + 2 + np.random.randn(m, 1)
```

```python
plt.scatter(X, y)
plt.show()
```

```python
X[0]
```

```python
poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_new = poly_features.fit_transform(X)
X_new[0]
```

```python
lin_reg = LinearRegression().fit(X_new, y)
lin_reg.intercept_, lin_reg.coef_
```

```python
def plot_train_val_error(model, X, y):
    X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2)
    train_error = []
    val_error = []
    for m in range(1, len(X_train)):
        model.fit(X_train[:m], y_train[:m])
        y_train_predict = model.predict(X_train[:m])
        y_valid_predict = model.predict(X_val)
        
        train_error.append(mean_squared_error(y_train[:m], y_train_predict))
        val_error.append(mean_squared_error(y_val, y_valid_predict))
        
    plt.figure(figsize=(8, 6))
    plt.plot(np.sqrt(train_error), '-+', color='salmon', label='train error')
    plt.plot(np.sqrt(val_error), '-', color='blue', label='valid_error');
    plt.legend()
        
plot_train_val_error(lin_reg, X, y)
```

```python
poly_feat = Pipeline([
    ('scaling', PolynomialFeatures(degree=10, include_bias=False)),
    ('model', LinearRegression())
])

plot_train_val_error(poly_feat, X, y)
```

# Ridge Regression

```python
ridge = Ridge(alpha=1, solver='cholesky').fit(X, y)
ridge.predict([[1.5]])
```

```python
sgd_reg = SGDRegressor(penalty='l2').fit(X,y)
sgd_reg.predict([[1.5]])
```

# Lasso Regression

```python
lasso = Lasso(alpha=0.1).fit(X, y)
lasso.predict([[1.5]])
```

```python
lsgd = SGDRegressor(penalty='l1').fit(X,y)
lsgd.predict([[1.5]])
```

# ElasticNet

```python
elastic_net = ElasticNet(alpha=0.1, l1_ratio=0.5)
elastic_net.fit(X, y)
elastic_net.predict([[1.5]])
```
