```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import fetch_openml
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from sklearn.manifold import TSNE
from sklearn.metrics import accuracy_score
```

# Loading the data set

- Building a classifier on the original data set

```python
mnist = fetch_openml(name='mnist_784', version=1)
X_train, X_test = mnist.data[:60000], mnist.data[60000:]
y_train, y_test = mnist.target[:60000], mnist.target[60000:]
```

# Building Models

Exercise 1:

Load the MNIST dataset (introduced in Chapter 3) and split it into a training set
and a test set (take the first 60,000 instances for training, and the remaining
10,000 for testing). Train a Random Forest classifier on the dataset and time how
long it takes, then evaluate the resulting model on the test set. Next, use PCA to
reduce the dataset’s dimensionality, with an explained variance ratio of 95%.
Train a new Random Forest classifier on the reduced dataset and see how long it
takes. Was training much faster? Next, evaluate the classifier on the test set. How
does it compare to the previous classifier?

```python
rf_clf = RandomForestClassifier()
%time rf_clf.fit(X_train, y_train)
```

```python
# Evaluating the model on the test set
y_pred = rf_clf.predict(X_test)
score = accuracy_score(y_test, y_pred)
print(f'Test score accuracy : {score:.3f}')
```

- Reducing the data set to a convarience of 95% and building a model with it

```python
pca = PCA(n_components=0.95)
X_train_reduced = pca.fit_transform(X_train)
X_test_reduced = pca.transform(X_test)
```

```python
%time rf_clf.fit(X_train_reduced, y_train)
```

```python
X_train_reduced.shape, X_test_reduced.shape
```

```python
y_pred = rf_clf.predict(X_test_reduced)
reduced_score = accuracy_score(y_test, y_pred)
print(f'Test accuracy on reduced data set {reduced_score:.3f}')
```

- Applying PCA to the data set reduced both the training time and it accuracy score

Use t-SNE to reduce the MNIST dataset down to two dimensions and plot the

```python
m = np.random.permutation(10000)
tsne = TSNE(n_components=2, random_state=7)
mnist_reduced = tsne.fit_transform(mnist['data'][m])
```

```python
colors = ["#476A2A", "#7851B8", "#BD3430", "#4A2D4E", "#875525",
          "#A83683", "#4E655E", "#853541", "#3A3120", "#535D8E"]
plt.figure(figsize=(10, 10))
plt.xlim(mnist_reduced[:, 0].min(), mnist_reduced[:, 0].max())
plt.ylim(mnist_reduced[:, 1].min(), mnist_reduced[:, 1].max())

for i in range(len(mnist_reduced)):
    # actually plot the digits as text instead of using scatter
    plt.text(mnist_reduced[i, 0], mnist_reduced[i, 1], str(mnist.target[i]),
             fontdict={'weight': 'bold', 'size': 9})
plt.xlabel("First principal component")
plt.ylabel("Second principal component")
```
