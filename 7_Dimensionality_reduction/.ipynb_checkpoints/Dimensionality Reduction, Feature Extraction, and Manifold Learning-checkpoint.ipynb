# Dimensionality Reduction, Feature Extraction, and Manifold Learning

## Imports

```python
import mglearn
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_breast_cancer, fetch_lfw_people
from sklearn.decomposition import PCA

sns.set_style('darkgrid')
```

## Concept

```python
cancer = load_breast_cancer()
```

```python
mglearn.plots.plot_pca_illustration()
```

```python
fig, axes = plt.subplots(15, 2, figsize=(10, 20))
malignant = cancer.data[cancer.target == 0]
benign = cancer.data[cancer.target == 1]
ax = axes.ravel()
for i in range(30):
    _, bins = np.histogram(cancer.data[:, i], bins=50)
    ax[i].hist(malignant[:, i], bins=bins, color=mglearn.cm3(0), alpha=.5)
    ax[i].hist(benign[:, i], bins=bins, color=mglearn.cm3(2), alpha=.5)
    ax[i].set_title(cancer.feature_names[i])
    ax[i].set_yticks(())
ax[0].set_xlabel("Feature magnitude")
ax[0].set_ylabel("Frequency")
ax[0].legend(["malignant", "benign"], loc="best")
fig.tight_layout()
```

## Using PCA For Data Visualization 

```python
scaler =  StandardScaler()
X_scaled = scaler.fit_transform(cancer['data'])

pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X_scaled)
```

```python
print(f'Original Shape : {X_scaled.shape}')
print(f'New shape : {X_reduced.shape}')
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(X_reduced[:, 0], X_reduced[:, 1], hue=cancer['target'])
plt.xlabel('first component')
plt.ylabel('second component')
plt.show()
```

```python
pca.components_.shape
```

```python
pca.components_.ravel()
```

```python
pca_mat = pca.components_
plt.matshow(pca_mat, cmap='viridis')
plt.xticks(range(len(cancer.feature_names)), labels=cancer.feature_names, rotation=60, ha='left')
plt.colormaps()
plt.colorbar()
plt.show()
```

## Using PCA For Feature Extraction 

```python
people = fetch_lfw_people(min_faces_per_person=20, resize=0.7)
```

```python
people.keys()
```

```python
image_shape = people.images[0].shape
```

```python
fix, axes = plt.subplots(2, 5, figsize=(15, 8),
                         subplot_kw={'xticks': (), 'yticks': ()})
for target, image, ax in zip(people.target, people.images, axes.ravel()):
    ax.imshow(image, cmap='gray')
    ax.set_title(people.target_names[target])
```

```python
print(people.images.shape)
print(people.target_names.__len__())
```

```python
counts = np.bincount(people.target)
for i , (count, name) in enumerate(zip(counts, people.target_names)):
    print(f'{name} : {count}')
    if (i + 1) % 3 ==0 :
        print()
```

```python
mask = np.zeros(people.target.shape, dtype=np.bool)
for target in np.unique(people.target):
    mask[np.where(people.target == target)[0][:50]] = 1
X_people = people.data[mask]
y_people = people.target[mask]
# scale the grayscale values to be between 0 and 1
# instead of 0 and 255 for better numeric stability
X_people = X_people / 255.
```

```python
X_train, X_test, y_train, y_test =  train_test_split(X_people, y_people, 
                                                     stratify=y_people, random_state=0)
knn = KNeighborsClassifier(n_neighbors=1).fit(X_train , y_train)
knn.score(X_test, y_test)
```

```python
mglearn.plots.plot_pca_whitening()
```

```python
pca = PCA(n_components=100, whiten=True, random_state=0)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

print(f'shape of X_train_pca : {X_train_pca.shape}')
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train_pca, y_train)
print("Test set accuracy: {:.2f}".format(knn.score(X_test_pca, y_test)))
```

```python
print("pca.components_.shape: {}".format(pca.components_.shape))
```

```python
fix, axes = plt.subplots(3, 5, figsize=(15, 12),
                         subplot_kw={'xticks': (), 'yticks': ()})
for i, (component, ax) in enumerate(zip(pca.components_, axes.ravel())):
    ax.imshow(component.reshape(image_shape),
              cmap='viridis')
    ax.set_title("{}. component".format((i + 1)))
```

```python
mglearn.plots.plot_pca_faces(X_train, X_test, image_shape)
```

```python
mglearn.discrete_scatter(X_train_pca[:, 0], X_train_pca[:, 1], y_train)
plt.xlabel("First principal component")
plt.ylabel("Second principal component")   
```
